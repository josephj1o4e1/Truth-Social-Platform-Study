Import dataset using pandas. 

```python
import pandas as pd
DATA_PATH = "./Dataset/truth_social"
def load_users_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "users.tsv")
    return pd.read_table(tsv_path)
def load_follows_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "follows.tsv")
    return pd.read_table(tsv_path)
def load_truths_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "truths.tsv")
    return pd.read_table(tsv_path)
def load_quotes_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "quotes.tsv")
    return pd.read_table(tsv_path)
def load_replies_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "replies.tsv")
    return pd.read_table(tsv_path)
def load_media_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "media.tsv")
    return pd.read_table(tsv_path)
def load_hashtags_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "hashtags.tsv")
    return pd.read_table(tsv_path)
def load_externalurls_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "externalurls.tsv")
    return pd.read_table(tsv_path)
def load_truthhashtagEdges_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "truth_hashtag_edges.tsv")
    return pd.read_table(tsv_path)
def load_truthmediaEdges_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "truth_media_edges.tsv")
    return pd.read_table(tsv_path)
def load_truthexturlEdges_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "truth_external_url_edges.tsv")
    return pd.read_table(tsv_path)
def load_truthusertagEdges_data(data_path=DATA_PATH):
    tsv_path = os.path.join(data_path, "truth_user_tag_edges.tsv")
    return pd.read_table(tsv_path)
```
